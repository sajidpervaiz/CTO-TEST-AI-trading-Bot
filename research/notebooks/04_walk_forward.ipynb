# 04 — Walk-Forward Optimization

Validate strategy robustness using out-of-sample walk-forward analysis.

```python
import sys
sys.path.insert(0, '../../')

from research.utils.data_loader import load_candles
from research.utils.feature_engineering import add_all_indicators
from research.utils.backtest_runner import walk_forward_analysis, monte_carlo_simulation
from research.utils.visualization import equity_curve

df = load_candles('binance', 'BTC/USDT:USDT', '4h', limit=10000)
df = add_all_indicators(df)
print(f'Dataset: {len(df):,} candles')
```

```python
STRATEGY  = 'ema_crossover'
N_SPLITS  = 5

wfo_df = walk_forward_analysis(df, STRATEGY, n_splits=N_SPLITS)
print(f'Walk-Forward Results ({N_SPLITS} splits):')
wfo_df
```

```python
import pandas as pd
print('Aggregate WFO metrics:')
print(wfo_df[['total_return','sharpe_ratio','max_drawdown','win_rate']].describe())
```

```python
returns_series = df['close'].pct_change().dropna()
mc_df = monte_carlo_simulation(returns_series, n_simulations=500, n_periods=252)
print(f'Monte Carlo ({len(mc_df):,} sims):')
mc_df.describe()
```
