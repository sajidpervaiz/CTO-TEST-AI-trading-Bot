# 02 — Feature Analysis

Compute technical indicators and analyse feature correlations with forward returns.

```python
import sys
sys.path.insert(0, '../../')

from research.utils.data_loader import load_candles
from research.utils.feature_engineering import (
    add_all_indicators, add_returns, add_volatility_features, compute_feature_importance
)
from research.utils.visualization import correlation_heatmap, indicator_chart

df = load_candles('binance', 'BTC/USDT:USDT', '1h', limit=5000)
df = add_all_indicators(df)
df = add_returns(df)
df = add_volatility_features(df)

print(f'DataFrame shape: {df.shape}')
df.tail()
```

```python
fig = indicator_chart(df.tail(200), ['rsi_14', 'stoch_k', 'stoch_d'], title='Oscillators')
fig.show()
```

```python
feature_cols = ['rsi_14', 'macd', 'bb_pct', 'atr_pct', 'rvol_20', 'ret_1', 'ret_5', 'ret_20']
fig = correlation_heatmap(df[feature_cols].dropna(), title='Feature Correlation Matrix')
fig.show()
```

```python
importance = compute_feature_importance(df, target_col='ret_1', top_n=20)
print('Top 20 features by LightGBM importance:')
importance
```
