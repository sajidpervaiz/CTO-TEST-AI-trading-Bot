# 01 — Data Exploration

Load and visualise historical OHLCV data from the TimescaleDB storage layer.

```python
import sys
sys.path.insert(0, '../../')

import pandas as pd
import numpy as np
from research.utils.data_loader import load_candles
from research.utils.visualization import candlestick_chart

EXCHANGE = 'binance'
SYMBOL   = 'BTC/USDT:USDT'
TF       = '15m'

print(f'Loading {SYMBOL} {TF} candles from {EXCHANGE}…')
df = load_candles(EXCHANGE, SYMBOL, TF, limit=2000)
print(f'Loaded {len(df):,} candles  |  {df.index[0]} → {df.index[-1]}')
df.head()
```

```python
df.describe()
```

```python
fig = candlestick_chart(
    df.tail(500),
    title=f'{SYMBOL} — {TF} Candlestick',
    show_volume=True,
)
fig.show()
```

```python
returns = df['close'].pct_change().dropna()
print(f'Mean return : {returns.mean():.4%}')
print(f'Std  return : {returns.std():.4%}')
print(f'Sharpe (ann): {returns.mean() / returns.std() * (365*24*4)**0.5:.2f}')
returns.hist(bins=100, figsize=(12,4))
```
