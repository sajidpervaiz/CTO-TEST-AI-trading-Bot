# 03 — Signal Research

Backtest registered trading strategies and compare performance metrics.

```python
import sys
sys.path.insert(0, '../../')

from research.utils.data_loader import load_candles
from research.utils.feature_engineering import add_all_indicators
from research.utils.backtest_runner import run_strategy
from research.utils.visualization import equity_curve
from engine.strategy_registry import registry

print('Available strategies:', registry.list())
```

```python
df = load_candles('binance', 'BTC/USDT:USDT', '1h', limit=5000)
df = add_all_indicators(df)

results = {}
for name in registry.list():
    try:
        r = run_strategy(df, name, symbol='BTC/USDT')
        results[name] = r
        print(f'{name:<25}  return={r.total_return:+.2%}  sharpe={r.sharpe_ratio:+.2f}  dd={r.max_drawdown:.2%}  trades={r.num_trades}')
    except Exception as e:
        print(f'{name:<25}  ERROR: {e}')
```

```python
import pandas as pd
comparison = pd.DataFrame({
    name: result.to_dict()
    for name, result in results.items()
}).T
comparison.sort_values('sharpe_ratio', ascending=False)
```
